```python
import pandas
```

```python
data = pandas.read_excel("C:\\Users\\Agne\\Desktop\\PythonCourse\\Homework\\finalproject\\P_Data_Extract_From_Gender_Statistics (2).xlsx")
data
```

```python
data = data.drop(index=[116,117,118,119,120])
data
```

```python
data = data.drop(columns='Series Code')
data = data.round(1)
data = data.loc[data["2015 [YR2015]"] != 0.0 ]
data
```

```python
data1 = data.pivot(index=['Country Name', 'Country Code'],
                       columns=['Series Name'], values=['2015 [YR2015]','2016 [YR2016]','2017 [YR2017]'])
data1
```

```python
data2 = data1.dropna()
```

```python
data2
```

```python
data2.loc['Mean'] = data2.mean()
data2
```

```python
data_stem = data2[[('2015 [YR2015]','Female share of graduates from Science, Technology, Engineering and Mathematics (STEM) programmes, tertiary (%)'),
                   ('2016 [YR2016]','Female share of graduates from Science, Technology, Engineering and Mathematics (STEM) programmes, tertiary (%)'),
                  ('2017 [YR2017]','Female share of graduates from Science, Technology, Engineering and Mathematics (STEM) programmes, tertiary (%)')]]
```

```python
data_stem
```

```python
data_stem_mean = data_stem.mean(axis=1)
data_stem_std = data_stem.std(axis=1)
```

```python
data_stem_mean.to_list()
```

```python
data_gdp = data2[[('2015 [YR2015]','GDP per capita (Current US$)'),
                   ('2016 [YR2016]','GDP per capita (Current US$)'),
                  ('2017 [YR2017]','GDP per capita (Current US$)')]]
```

```python
data_gdp
```

```python
data_gdp_mean = data_gdp.mean(axis=1)
```

```python
data_gdp_std = data_gdp.std(axis=1)
```

```python
data_gdp_mean.to_list()
```

```python
data2["Mean of female share of graduates from STEM programmes, tertiary (%)"] = data_stem_mean.to_list()
data2["Mean of GDP per capita (Current US$)"] = data_gdp_mean.to_list()
```

```python
data2
```

# Data visualisation

```python
import matplotlib.pyplot as pl
```

### Females in stem data

```python
years = ["2015","2016","2017"]
meanstem = data_stem.iloc[len(data_stem.index)-1]
meanstem = meanstem.to_list()
```

```python
pl.bar(years,meanstem)
pl.ylim(33,35)
pl.ylabel("Mean of female share of graduates from STEM programmes, tertiary (%)")
pl.xlabel("Year")
```

```python
pl.hist(data2["Mean of female share of graduates from STEM programmes, tertiary (%)"],bins=20)
pl.ylabel("Number of countries")
pl.xlabel("Mean of female share of graduates from STEM programmes, tertiary (%)")
```
